```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.transform import resize
import os

# Paths
csv_path = '/content/drive/MyDrive/mini_proj_data/HAM10000/hmnist_28_28_RGB.csv'
image_dir = '/content/drive/MyDrive/mini_proj_data/HAM10000/HAM10000_images_part_1'
meta_path = '/content/drive/MyDrive/mini_proj_data/HAM10000/HAM10000_metadata.csv'

# Load CSV and metadata
df = pd.read_csv(csv_path)
meta = pd.read_csv(meta_path)

# Pick any image index (e.g., 0)
idx = 0

# Extract pixel vector and reshape to 28x28x3
pixels = df.iloc[idx, :-1].values  # exclude label column
img_csv = pixels.reshape(28, 28, 3).astype(np.uint8)

# Find the corresponding original image (same lesion type for visual check)
# We'll just randomly pick the first image of that label type
label = df.iloc[idx, -1]
sample_image_id = meta[meta['dx'] == meta['dx'].unique()[label]]['image_id'].iloc[0]
img_path = os.path.join(image_dir, f"{sample_image_id}.jpg")

# Load and resize the original image
img_orig = imread(img_path)
img_resized = resize(img_orig, (28, 28))
img_resized = (img_resized * 255).astype(np.uint8)

# Calculate absolute difference
diff = np.abs(img_resized.astype(np.int16) - img_csv.astype(np.int16))
mean_diff = np.mean(diff)

# Show results
plt.figure(figsize=(10, 4))
plt.subplot(1, 3, 1)
plt.imshow(img_orig)
plt.title("Original (Full Res)")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(img_resized)
plt.title("Original Resized to 28x28")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(img_csv)
plt.title("CSV Image (From hmnist_28_28_RGB)")
plt.axis("off")

plt.suptitle(f"Mean Pixel Difference: {mean_diff:.2f}", fontsize=14)
plt.show()
```
